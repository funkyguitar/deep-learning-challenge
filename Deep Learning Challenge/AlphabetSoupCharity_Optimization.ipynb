```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
applicationDropDF = application_df.drop(['EIN', 'NAME', 'AFFILIATION'], axis=1)
applicationDropDF
```

```python
# Determine the number of unique values in each column.
n = applicationDropDF.nunique(axis=0)
  
print("No.of.unique values in each column :\n",
      n)
```

```python
# Look at APPLICATION_TYPE value counts for binning
applicationDropDF.APPLICATION_TYPE.value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
# Replace in dataframe
for app in applicationDropDF:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
applicationDropDF.CLASSIFICATION.value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
applicationDropDF.groupby('CLASSIFICATION').filter(lambda x: len(x) >= 2).CLASSIFICATION.value_counts()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
# Define cutoff value
cutoff = 10

# Create a list of classifications with count less than the cutoff
class_counts = application_df['CLASSIFICATION'].value_counts()
classifications_to_replace = list(class_counts[class_counts < cutoff].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
with_dummies_df = pd.get_dummies(applicationDropDF)
with_dummies_df
```

```python
# Split our preprocessed data into our features and target arrays
y = with_dummies_df["IS_SUCCESSFUL"].values
X = with_dummies_df.drop(["IS_SUCCESSFUL"],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 140
hidden_nodes_layer2 = 50

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"]) 
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=120) 
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save('AlphabetSoupCharity_Optimization.h5')
```

# **Predicting Successful Applicants Who Receive Funding From Alphabet Soup**

Using neural networks, I have developed a tool that can help identify which applicants who seek funding have the best chance of success in their ventures. 


**Defining Target and Feature Variables**

The target variable is the model_accuracy. The feature variables include the columns, bins, epochs, neurons, and layers. The EIN and NAME were removed from the analysis since they are identification columns and are neither targets nor features.


**Optimization Methods**

In my first attempt at optimizing the model, I dropped the AFFILIATION column and increased hidden nodes layer1 to from 80 to 120, and hidden nodes layer2 from 30 to 45. This had a negligible impact, however. Since adding more layers does not guarantee more accurate results, I was curious to experiment with reducing the number of layers, and reduced hidden nodes layer1 to 40, and hidden nodes layer2 to 20. This unfortunately reduced accuracy by 9%, from 73% to 66% . Finally, I increased hidden nodes layer1 to 140, and hidden nodes layer2 to 50, and also increased the number of epochs from 100 to 120. This also had a 66% prediction accuracy.


**Ideas For Better Optimization**

Some ideas for improving predictions in future optimization attempts include dropping more columns, creating more bins for rare occurrences in columns, and increasing or decreasing the number of values for each bin.
